```python
# 4.3
print("Hello World!")
img = cv2.imread("lena.png")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Lena Color")
```

```python
plt.imshow(gray, cmap = 'gray')
plt.title("Lena Gray")
```

```python
# 5.1
print(img.shape)
print(img.size)
print(img.dtype)
```

```python
# 5.2: Flipping images
# Lena Color Horizontal Flip
h_flip_rgb = np.zeros(img.shape, np.uint8)
for row in range (0, img.shape[0]-1):
    for column in range(0, img.shape[1]-1):
        h_flip_rgb[row][column] = img[row][img.shape[1] -1 - column]
plt.figure()
plt.imshow(cv2.cvtColor(h_flip_rgb, cv2.COLOR_BGR2RGB))
plt.title("Lena Color horizontal flip")

# Lena Color verical flip
v_flip_rgb = np.zeros(img.shape, np.uint8)
for row in range (0, img.shape[0]-1):
    for column in range(0, img.shape[1]-1):
        v_flip_rgb[row][column] = img[img.shape[0] -1 -row][column]
plt.figure()
plt.imshow(cv2.cvtColor(v_flip_rgb, cv2.COLOR_BGR2RGB))
plt.title("Lena Color verical flip")

# Lena GrayScale Horizontal Flip
h_flip_gray = np.zeros(gray.shape, np.uint8)
for row in range (0, gray.shape[0]-1):
    for column in range(0, gray.shape[1]-1):
        h_flip_gray[row][column] = gray[row][gray.shape[1] -1 - column]
plt.figure()
plt.imshow(h_flip_gray, cmap = 'gray')
plt.title("Lena Gray horizontal flip")

# Lena GrayScale verical flip
v_flip_gray = np.zeros(gray.shape, np.uint8)
for row in range (0, gray.shape[0]-1):
    for column in range(0, gray.shape[1]-1):
        v_flip_gray[row][column] = gray[gray.shape[0] -1 -row][column]
plt.figure()
plt.imshow(v_flip_gray, cmap = 'gray')
plt.title("Lena Gray verical flip")
```

```python
# 5.3 Negative Image Pixels

# Negative Color
n_color = np.zeros(img.shape, np.uint8)
for row in range (0, img.shape[0]-1):
    for column in range(0, img.shape[1]-1):
        for color in range(0, img.shape[2]):
            n_color[row][column][color] = 255 - img[row][column][color]
plt.figure()
plt.imshow(cv2.cvtColor(n_color, cv2.COLOR_BGR2RGB))
plt.title("Lena Color Negative")

# Negative GrayScale
n_gray = np.zeros(gray.shape, np.uint8)
for row in range (0, gray.shape[0]-1):
    for column in range(0, gray.shape[1]-1):
            n_gray[row][column] = 255 - gray[row][column]
plt.figure()
plt.imshow(n_gray, cmap = 'gray')
plt.title("Lena Gray Negative")
```

```python
# 5.4 Separating Color Channels

# Create 3 empty images
filtered_images = [np.zeros(img.shape, np.uint8), np.zeros(img.shape, np.uint8), np.zeros(img.shape, np.uint8)]
for row in range (0, img.shape[0]-1):
    for column in range(0, img.shape[1]-1):
        for color in range(0, img.shape[2]):
            # Assign the value of the color to its matching image
            filtered_images[color][row][column][color] = img[row][column][color]

# Plot all images
img_titles = ["blue", "green", "red"]
for filtered_image, image_color in zip(filtered_images, img_titles):
    plt.figure()
    plt.imshow(cv2.cvtColor(filtered_image, cv2.COLOR_BGR2RGB))
    plt.title("Lena " + image_color)
```

```python
# Merging the separate outputs to obtain the same original picture
merged = sum(filtered_images)
plt.figure()
plt.imshow(cv2.cvtColor(merged, cv2.COLOR_BGR2RGB))
plt.title("Lena merged ")
```
